# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count.

ax = sns.countplot(x = df['Gender'])

for bars in ax.containers:
     ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

ax = df.groupby(by = df['Gender'])[['Amount']].sum().sort_values(by = 'Amount',ascending = False).plot(kind = 'bar')
```

### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.

## Age

```python
ax = sns.countplot( x = df['Age Group'], hue = df['Gender'])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group.

age_amount= df.groupby(by = df['Age Group'])[['Amount']].sum().sort_values(by = 'Amount',ascending = False).plot(kind = 'bar')
```

### From above graphs we can see that most of the buyers are of age group between 26-35 yrs female.


## States

```python
# total number of orders from top 5 states
State_order = df.groupby(by = df['State'])[['Orders']].sum().sort_values(by = 'Orders',ascending = False).head(5).plot(kind = 'bar')
```

```python
# total amount/sales from top 5 states

sales_state = df.groupby(by =df['State'])[['Amount']].sum().sort_values(by='Amount', ascending=False).head(5)

sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.

## Marital_Status

```python
 ax = sns.countplot( x =df['Marital_Status'])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 sale_status =df.groupby(by=['Marital_Status', 'Gender'])[['Amount']].sum().sort_values(by = 'Amount',ascending = False)
 fig = plt.figure(figsize = (6,5))                                                                                       
sns.barplot(data = sale_status, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

##  Occupation

```python
ax = df['Occupation'].value_counts().head(5).plot(kind = 'bar')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python

ax = df.groupby(by = df['Occupation'])[['Amount']].sum().sort_values(by = 'Amount',ascending = False).head(5).plot(kind = 'bar')
```

### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product_Category

```python
product_c = df.groupby(by = df['Product_Category'])[['Amount']].sum().sort_values(by = 'Amount',ascending = False).head(5).plot(kind = 'bar')
```

```python
 ax= df.groupby(by = df['Product_Category'])[['Orders']].sum().sort_values(by = 'Orders',ascending = False).head(5).plot(kind='bar')
for bars in ax.containers:
    ax.bar_label(bars)
```

### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

## Product_ID

```python
# top 10 most sold products (same thing as above)
product_Orders = df.groupby(by = df['Product_ID'])[['Orders']].sum().sort_values(by = 'Orders',ascending = False).head(5).plot(kind = 'bar')
```
